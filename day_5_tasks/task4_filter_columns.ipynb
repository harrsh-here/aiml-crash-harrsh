```python
import pandas as pd
```

```python
df = pd.read_csv("learners.csv")
df.head()
```

```python
subset = df[["name", "role", "score"]]
subset.sample()
```

```python
filtered = subset[df["role"] == "Intern"]
print("Filter: role == Intern") # this filter is printing only those values in which learners are intern
filtered
```

```python
filtered2 = subset[df["score"] >= 85]
print("\nFilter: score >= 85") # THis filter is printing rows where score is 85 or more
filtered2
```
