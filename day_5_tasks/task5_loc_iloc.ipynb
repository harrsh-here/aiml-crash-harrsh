```python
import pandas as pd
```

```python
df = pd.read_csv("learners.csv")
```

### .loc selects by label (index label and column name)

```python
df.loc[0:2, ["name", "score"]]
```

### .iloc selects by integer position (row number and column number)

```python
df.iloc[0:3, [0, 2]]
```

    -- .loc uses labels so the range is inclusive on both ends
    -- .iloc uses positions like Python slices so the end is exclusive
