```python
import pandas as pd
```

```python
df = pd.read_csv("learners.csv")
```

# Missing Values Per Columns

```python
df.isnull().sum()
```

# After dropna

```python
df_dropped = df.dropna()
```

```python
df_dropped.isnull().sum()
```

# fillna

```python
df_filled = df.copy()
df_filled["score"] = df_filled["score"].fillna(df_filled["score"].mean())
```

```python
df_filled.isnull().sum()
```
