```python
def build_profile_card(profile: dict) -> str:
    line1 = f"Name   : {profile['name']}"
    line2 = f"Role   : {profile['role']}"
    line3 = f"Skills : {', '.join(profile['skills'])}"
    line4 = f"Batch  : {profile['batch']}"
    return f"{line1}\n{line2}\n{line3}\n{line4}"
```

```python
name = "Harsh Patidar"
role = "Intern"
skills = ["Python", "Git", "Pandas"]
batch = "2025-B"
```

```python
profile = {
    "name": name,
    "role": role,
    "skills": skills,
    "batch": batch,
}
```

```python
print(build_profile_card(profile))
```
