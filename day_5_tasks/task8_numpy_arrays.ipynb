```python
import numpy as np
```

```python
scores = np.array([95, 88, 82, 76, 91, 85, 70, 78, 88])
scores
```

```python
print(f"shape={scores.shape}  dtype={scores.dtype}  ndim={scores.ndim}")
```

```python
ranks = np.arange(1, 10)
```

```python
print(f"shape={ranks.shape}  dtype={ranks.dtype}  ndim={ranks.ndim}")
```

```python
zeros_board = np.zeros((2, 4))
zeros_board
```

```python
print(f"shape={zeros_board.shape}  dtype={zeros_board.dtype}  ndim={zeros_board.ndim}")
```

```python
percentiles = np.linspace(0, 100, 5)
percentiles
```

```python
print(f"shape={percentiles.shape}  dtype={percentiles.dtype}  ndim={percentiles.ndim}")
```
