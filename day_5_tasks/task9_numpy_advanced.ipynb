# Similarity Calculation 

```python
def cosine_similarity(v1: np.ndarray, v2: np.ndarray) -> float:
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
harsh_profile = np.array([1, 2, 3])
gaurav_profile = np.array([4, 5, 6])
dheeraj_profile = np.array([1, 0, 0])
priya_profile = np.array([0, 1, 0])
```

```python
print(f"\ncosine_similarity(harsh_profile, gaurav_profile) = {cosine_similarity(harsh_profile, gaurav_profile):.4f}")
```

```python
print(f"cosine_similarity(dheeraj_profile, priya_profile) = {cosine_similarity(dheeraj_profile, priya_profile):.4f}")
```
