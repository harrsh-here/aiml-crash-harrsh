```python
class Learner:
    def __init__(self, name: str, batch: str, project: str):
        self.name = name
        self.batch = batch
        self.project = project

    def summary(self) -> str:
        return f"Learner: {self.name} | Batch: {self.batch} | Project: {self.project}"
```

```python
learner = Learner("Harsh Patidar", "2025-B", "Sales Dashboard")
```

```python
print(learner.summary())
```
