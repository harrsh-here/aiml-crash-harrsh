# Task 1 - Derivative and Slope Intuition

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 2*x + 1
```

```python
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)
```

choosing 3 points to compute slope at

```python
x1 = -3
numerical_derivative(f, x1)
```

```python
x2 = -1
numerical_derivative(f, x2)
```

```python
x3 = 2
numerical_derivative(f, x3)
```

plotting the curve with tangent lines at all 3 points

```python
x = np.linspace(-5, 4, 200)
y = f(x)

points = [x1, x2, x3]
colors = ["red", "green", "blue"]

plt.figure(figsize=(9, 6))
plt.plot(x, y, "k-", linewidth=2, label="f(x) = x² + 2x + 1")

for xi, color in zip(points, colors):
    slope = numerical_derivative(f, xi)
    yi = f(xi)
    tangent_x = np.linspace(xi - 1.5, xi + 1.5, 50)
    tangent_y = slope * (tangent_x - xi) + yi
    plt.plot(tangent_x, tangent_y, color=color, linestyle="--", label=f"tangent at x={xi} (slope={slope:.2f})")
    plt.plot(xi, yi, "o", color=color, markersize=8)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = x² + 2x + 1 with Tangent Lines")
plt.legend()
plt.grid(True)
plt.show()
```

### 3 Observations

1. At x=-3 the slope is -4 (negative), which means the function is going downhill. In ML this tells gradient descent to move right (increase the parameter) to reduce loss.

2. At x=-1 the slope is basically 0, this is the minimum of the curve. In optimization this is where gradient descent wants to reach because the loss cant go any lower.

3. At x=2 the slope is 6 (positive), the function is going uphill. Gradient descent would move left (decrease the parameter) to get closer to the minimum.
