# Task 3 - Probability Basics with Sampled Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

generating 1000 samples from uniform and normal distributions

```python
np.random.seed(42)
uniform_samples = np.random.uniform(0, 10, 1000)
normal_samples = np.random.normal(5, 2, 1000)
```

### Uniform distribution stats

```python
uniform_samples.mean()
```

```python
uniform_samples.std()
```

### Normal distribution stats

```python
normal_samples.mean()
```

```python
normal_samples.std()
```

### Histograms side by side

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(uniform_samples, bins=25, color="skyblue", edgecolor="black")
axes[0].set_title("Uniform Distribution")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

axes[1].hist(normal_samples, bins=25, color="salmon", edgecolor="black")
axes[1].set_title("Normal Distribution")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### Interpretation

The uniform histogram is roughly flat because every value between 0 and 10 has equal chance of showing up. The normal histogram has a bell shape where most values cluster around the mean (5) and fewer values appear as we go further away. In ML we see normal distributions a lot in things like weight initialization and error distributions.
