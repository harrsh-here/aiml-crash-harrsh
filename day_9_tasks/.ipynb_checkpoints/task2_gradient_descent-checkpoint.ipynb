# Task 2 - Gradient Descent Mini-Experiment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def loss(theta):
    return (theta - 3.5)**2 + 1
```

```python
def gradient(theta):
    return 2 * (theta - 3.5)
```

```python
def run_gd(lr, steps=12, start=10.0):
    theta = start
    history = []
    for i in range(steps):
        g = gradient(theta)
        l = loss(theta)
        history.append({"iteration": i+1, "theta": round(theta, 4), "loss": round(l, 4), "gradient": round(g, 4)})
        theta = theta - lr * g
    return pd.DataFrame(history)
```

### Standard learning rate (0.1)

```python
df_std = run_gd(lr=0.1)
df_std
```

### Low learning rate (0.01)

```python
df_low = run_gd(lr=0.01)
df_low
```

### High learning rate (0.9)

```python
df_high = run_gd(lr=0.9)
df_high
```

### Comparing all 3 loss curves

```python
plt.figure(figsize=(10, 5))
plt.plot(df_std["iteration"], df_std["loss"], marker="o", label="lr=0.1")
plt.plot(df_low["iteration"], df_low["loss"], marker="s", label="lr=0.01")
plt.plot(df_high["iteration"], df_high["loss"], marker="^", label="lr=0.9")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss Curve Comparison - Different Learning Rates")
plt.legend()
plt.grid(True)
plt.show()
```

### Comparison

lr=0.1 converges smoothly to the minimum in about 10 steps. lr=0.01 is too slow and barely moves in 12 steps. lr=0.9 overshoots and bounces around the minimum but still gets close. So picking the right learning rate really matters.
